# **Imágenes**

Las imágenes que necesito son las siguientes:
1. configuración: 86 imágenes. Son las manos
2. orientación: 64 imágenes. Son la orientación de las manos.
3. orientación modal derecha. 52 imágenes. Orientación de la mano derecha
4. orientación modal izquierda. 52 imágenes. Orientación de la mano izquierda.
5. boca. 34 imágenes.
6. cabeza. 8 imágenes.
7. cejas. 5 imágenes.
8. hombros. 9 imágenes.
9. mejillas. 4 imágenes.
10. ojos. 4 imágenes.

```python
# Enlaces de la configuración (86 imágenes)

# Creamos la lista para luego introducir los enlaces
listaConf=list()
listaNumC=list()

# hacemos un bucle for en el que el rango contiene las 86 imágenes
for cf in range(1, 87):
  # el número lo convertimos en string
  numbercf = str (cf)
  # insertamos el número para que se vayan modificando los enlaces
  urlconfig = 'http://lse-sign.bcbl.eu/web-busqueda/wp-content/themes/twentyeleven/img_bcbl/configuracion/'+numbercf+'.jpg'
  #agregamos los enlaces a la lista
  listaConf.append(urlconfig)
  listaNumC.append(numbercf)
# creo la tabla con el número y el enlace a la imagen
miDiccionario = {'Numero':listaNumC, 'Imagen': listaConf}
df = pd.DataFrame(miDiccionario, columns=['Numero','Imagen'])
# exporto a csv los datos
df.to_csv('ImagenesConfiguracion.csv')
# imprimo la tabla de datos
df
```

```python
# Enlaces de la orientación (64 imágenes)

# Seguimos los mismos pasos que anteriormente
listaOri = list()
listaNumO = list()
for ori in range(1, 65):
  numberori = str (ori)
  urlori = 'http://lse-sign.bcbl.eu/web-busqueda/wp-content/themes/twentyeleven/img_bcbl/configuracion/orientacion/'+numberori+'.jpg'
  listaOri.append(urlori)
  listaNumO.append(numberori)
# creo la tabla con el número y el enlace a la imagen
miDic2 = {'Numero':listaNumO, 'Imagen': listaOri}
df2 = pd.DataFrame(miDic2, columns=['Numero','Imagen'])
# exporto a csv los datos
df2.to_csv('ImagenesOrientacion.csv')
# imprimo la tabla de datos
df2
```

```python
# Enlaces de la orientación modal derecha (52 imágenes)

# Seguimos los mismos pasos que anteriormente
listaODcha = list()
listaNumODcha=list()
for od in range(1, 53):
  numberod = str (od)
  urlod = 'http://lse-sign.bcbl.eu/web-busqueda/wp-content/themes/twentyeleven/img_bcbl/configuracion/orientacion_modal/m_der/'+numberod+'.jpg'
  listaODcha.append(urlod)
  listaNumODcha.append(numberod)
# creo la tabla con el número y el enlace a la imagen
miDic3 = {'Numero':listaNumODcha, 'Imagen': listaODcha}
df3 = pd.DataFrame(miDic3, columns=['Numero','Imagen'])
# exporto a csv los datos
df3.to_csv('ImagenesOrientacionDcha.csv')
# imprimo la tabla de datos
df3
```

```python
# Enlaces de la orientación modal izquierda (52 imágenes)

# Seguimos los mismos pasos que anteriormente
listaOIzq = list()
listaNumOIzq=list()
for oi in range(1, 53):
  numberoi = str (oi)
  urloi = 'http://lse-sign.bcbl.eu/web-busqueda/wp-content/themes/twentyeleven/img_bcbl/configuracion/orientacion_modal/m_izq/'+numberoi+'.jpg'
  listaOIzq.append(urloi)
  listaNumOIzq.append(numberoi)
# creo la tabla con el número y el enlace a la imagen
miDic4 = {'Numero':listaNumOIzq, 'Imagen': listaOIzq}
df4 = pd.DataFrame(miDic4, columns=['Numero','Imagen'])
# exporto a csv los datos
df4.to_csv('ImagenesOrientacionIzq.csv')
# imprimo la tabla de datos
df4
```

```python
# Obtenemos los elementos no manuales (boca, cabeza, cejs, hombros, mejillas y ojos)  64 imágenes.

# Seguimos los mismos pasos que anteriormente
listaENM = list()
listaNombreENM = list()
# 1. Hacemos web scraping de la página que tiene todos los elementos manuales para así poder obtener los enlaces de las imágenes
r = urlopen("http://lse-sign.bcbl.eu/web-busqueda/wp-content/themes/twentyeleven/img_bcbl/elementosnomanuales/")
bs = BeautifulSoup(r.read(), "html.parser")
r.close()

for link in bs.find_all("a"):
  enm = link.get("href")
  # 2. Como la web contiene atributos href que no necesitamos utilizamos el if como filtro para que solo utilice las imágenes
  # que queremos (.jpg)
  if ".jpg" in enm:
    urlenm = 'http://lse-sign.bcbl.eu/web-busqueda/wp-content/themes/twentyeleven/img_bcbl/elementosnomanuales/'+enm
    listaENM.append(urlenm)
    listaNombreENM.append(enm)
# creo la tabla con el nombre y el enlace a la imagen
miDic5 = {'Nombre':listaNombreENM, 'Imagen': listaENM}
df5 = pd.DataFrame(miDic5, columns=['Nombre','Imagen'])
# exporto a csv los datos
df5.to_csv('ImagenesElemNoManual.csv')
# imprimo la tabla de datos
df5
```

# **VÍDEOS**

En cuanto a la obtención de los vídeos, vi que en una de estas páginas (1) se podían sacar todos los vídeos de la misma manera ya que en la web 2 están todos unidos diferenciados por números, por lo que voy a realizar un bucle para obtener los enlaces.

1. http://old.arasaac.org/videos_lse.php?Submit=Buscar&busqueda=basico&p=2%7C3%7C6&arbol=1&tipo_palabra=99&TXTlocate=1&letra=&filtrado=1&orden=desc
2. http://old.arasaac.org/repositorio/LSE_acepciones_mp4/

```python
# Creo dos listas, una para los enlaces y otra para los identificadores
listaVideo=list()
listaNumero=list()

# Hago un bucle que va desde el primer número de vídeo hasta el último
for i in range(12368, 23414):
  # number recoge los números
  number = str (i)
  # urlvideo recoge los enlaces
  urlvideo="http://old.arasaac.org/repositorio/LSE_acepciones_mp4/"+number+".mp4"
  # number y urlvideo se introduen en las listas creadas anteriormente
  listaVideo.append(urlvideo)
  listaNumero.append(number)
# Creo el dataframe
miDic = {'Numero':listaNumero, 'Video': listaVideo}
df = pd.DataFrame(miDic, columns=['Numero','Video'])
# exporto a csv los datos
df.to_csv('Videos.csv')
# imprimo la tabla de datos
df
```

# **Reducir la base de datos para utilizarla en el Chatbot**

```python
# Se abre el csv
df = pd.read_csv("/content/drive/MyDrive/TFG/CSV/TFG_Limpio.csv", delimiter=';', encoding="utf-8")
df
```

```python
# Se eliminan las filas cuyos lemas que contienen números (ej: abandonar1)
df=df[~df.Lema.str.contains('1')]
df=df[~df.Lema.str.contains('2')]
df=df[~df.Lema.str.contains('3')]
df=df[~df.Lema.str.contains('4')]
df=df[~df.Lema.str.contains('5')]
df=df[~df.Lema.str.contains('6')]
df=df[~df.Lema.str.contains('7')]
df=df[~df.Lema.str.contains('8')]
df=df[~df.Lema.str.contains('9')]
df=df[~df.Lema.str.contains('0')]
df
```

```python
# Se eliminan las filas que no tengan vídeos
df = df.drop(df[df['video']=='desconocido'].index)
df
```

```python
# La nueva tabla contiene las columnas Lema y numero (el identificador del vídeo)
tabla=df[['Lema', 'numero']]
tabla
```

```python
# Se renombran las columnas Lema como Query y numero como Response
tabla.rename(columns = {'Lema':'Query', 'numero':'Response'}, inplace = True)
tabla
```

```python
# Se exporta a csv
tabla.to_csv('TFG_Simple.csv', index=False)
```
